```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

LINESTYLE_UNIT = "-"
COLOR_UNIT = "black"
SCATTER_SIZE = 20.
OPACITY_SCATTER = 0.33
LINEWIDTH_MEAN = 1.
plt.rc('axes', labelsize=13.5)
plt.rcParams.update({'font.size': 11.0})
matplotlib.rc('xtick', labelsize=12.0) 
matplotlib.rc('ytick', labelsize=12.0) 
plt.rc('axes', labelsize=15.)
```

```python
rdf_mean_OO = np.load("CEA_256H2O_1ns_mean_OO.npy")
rdf_std_OO = np.load("CEA_256H2O_1ns_std_OO.npy")
rdf_mean_HH = np.load("CEA_256H2O_1ns_mean_HH.npy")
rdf_std_HH = np.load("CEA_256H2O_1ns_std_HH.npy")
rdf_grid = np.load("gridRDF_256H2O_1ns.npy")

plt.figure(dpi=300, figsize=(4.5,4))
plt.fill_between(rdf_grid[10:], rdf_mean_OO[10:]-rdf_std_OO[10:], rdf_mean_OO[10:]+rdf_std_OO[10:], color="r", linewidth=0., alpha=OPACITY_SCATTER, label="O-O")
plt.plot(rdf_grid[10:], rdf_mean_OO[10:], color="r", linestyle=LINESTYLE_UNIT, alpha=OPACITY_SCATTER,  linewidth=LINEWIDTH_MEAN) #label="O-O",
plt.fill_between(rdf_grid[10:], rdf_mean_HH[10:]-rdf_std_HH[10:], rdf_mean_HH[10:]+rdf_std_HH[10:], color="b", linewidth=0.,  alpha=OPACITY_SCATTER, label="H-H")
plt.plot(rdf_grid[10:], rdf_mean_HH[10:], color="b", linestyle=LINESTYLE_UNIT, alpha=OPACITY_SCATTER,  linewidth=LINEWIDTH_MEAN) #label="H-H",

plt.xlabel("r [Å]")
plt.ylabel("g(r)")

plt.legend()
plt.xlim(1.,6.)
plt.tight_layout()
plt.savefig("RDF_256H2O_300ps.pdf",dpi=300)
```
