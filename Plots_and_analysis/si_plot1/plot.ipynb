```python
rdf_val_64_mean = np.load("../H2O_write_UQ/CEA_64H2O_1ns_mean_OO.npy")
rdf_val_64_std = np.load("../H2O_write_UQ/CEA_64H2O_1ns_std_OO.npy")
rdf_val_256_mean = np.load("../H2O_write_UQ/CEA_256H2O_1ns_mean_OO.npy")
rdf_val_256_std = np.load("../H2O_write_UQ/CEA_256H2O_1ns_std_OO.npy")
rdf_bins = np.load("../H2O_write_UQ/gridRDF_256H2O_1ns.npy")
```

```python

fig, ax = plt.subplots(2,2, figsize=(10,7),dpi=300)



rdf_val_64_mean = np.load("../H2O_write_UQ/CEA_64H2O_1ns_mean_OO.npy")
rdf_val_64_std = np.load("../H2O_write_UQ/CEA_64H2O_1ns_std_OO.npy")
rdf_val_256_mean = np.load("../H2O_write_UQ/CEA_256H2O_1ns_mean_OO.npy")
rdf_val_256_std = np.load("../H2O_write_UQ/CEA_256H2O_1ns_std_OO.npy")
rdf_bins = np.load("../H2O_write_UQ/gridRDF_256H2O_1ns.npy")

ax[1,1].fill_between(rdf_bins[2:], rdf_val_256_mean[2:] - rdf_val_256_std[2:], rdf_val_256_mean[2:] + rdf_val_256_std[2:],color="tab:blue", alpha=0.5, linewidth=0.)
ax[1,0].fill_between(rdf_bins[2:], rdf_val_64_mean[2:] - rdf_val_64_std[2:], rdf_val_64_mean[2:] + rdf_val_64_std[2:],color="tab:blue", alpha=0.5, linewidth=0.)
ax[1,1].plot(rdf_bins[2:], rdf_val_256_mean[2:], label="256 H2O", color="tab:blue", linewidth=1.)
ax[1,0].plot(rdf_bins[2:], rdf_val_64_mean[2:], label="64 H2O", color="tab:blue", linewidth=1.)


ax[1,0].text(0.55, 0.95, 'O-O\n64 H$_2$O', transform=ax[1,0].transAxes, fontsize=14,
        verticalalignment='top')
ax[1,1].text(0.55, 0.95, 'O-O\n256 H$_2$O', transform=ax[1,1].transAxes, fontsize=14,
        verticalalignment='top')

#ax[1,1].plot(rdf_bins[2:], (rdf_val_64_std[2:])-(rdf_val_256_std[2:]), label="64 H2O", color="tab:blue")
#ax[1,1].plot(rdf_bins[2:], (rdf_val_64_mean[2:])-(rdf_val_256_mean[2:]), label="256 H2O", color="tab:green")

ax[1,1].set_xlabel("r [$\AA$]")
ax[1,0].set_xlabel("r [$\AA$]")
ax[1,0].set_ylabel("g(r)")
ax[0,0].set_ylabel("g(r)")

rdf_val_64_mean = np.load("../H2O_write_UQ/CEA_64H2O_1ns_mean_HH.npy")
rdf_val_64_std = np.load("../H2O_write_UQ/CEA_64H2O_1ns_std_HH.npy")
rdf_val_256_mean = np.load("../H2O_write_UQ/CEA_256H2O_1ns_mean_HH.npy")
rdf_val_256_std = np.load("../H2O_write_UQ/CEA_256H2O_1ns_std_HH.npy")
rdf_bins = np.load("../H2O_write_UQ/gridRDF_256H2O_1ns.npy")

ax[0,0].fill_between(rdf_bins[2:], rdf_val_64_mean[2:] - rdf_val_64_std[2:], rdf_val_64_mean[2:] + rdf_val_64_std[2:],color="tab:blue", alpha=0.5, linewidth=0.)
ax[0,1].fill_between(rdf_bins[2:], rdf_val_256_mean[2:] - rdf_val_256_std[2:], rdf_val_256_mean[2:] + rdf_val_256_std[2:],color="tab:blue", alpha=0.5, linewidth=0.)
ax[0,1].plot(rdf_bins[2:], rdf_val_256_mean[2:], label="256 H2O", color="tab:blue", linewidth=1.)
ax[0,0].plot(rdf_bins[2:], rdf_val_64_mean[2:], label="64 H2O", color="tab:blue", linewidth=1.)

ax[0,0].text(0.55, 0.95, 'H-H\n64 H$_2$O', transform=ax[0,0].transAxes, fontsize=14,
        verticalalignment='top')
ax[0,1].text(0.55, 0.95, 'H-H\n256 H$_2$O', transform=ax[0,1].transAxes, fontsize=14,
        verticalalignment='top')



fig.tight_layout()
#fig.show()
fig.savefig("rdf_64_256.pdf", dpi=300)
```
